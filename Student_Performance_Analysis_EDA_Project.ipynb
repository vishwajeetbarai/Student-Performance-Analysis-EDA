```python
# step1 : we will import the libraries 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
```

```python
# step2: load the datasets

df = pd.read_csv("C:/Users/rjcvi/Downloads/student_data.csv")
df.head()
```

```python
# step3 : we will understand the basics knowledge 
```

```python
df.shape
```

```python
df.info()
```

```python
# step4 : Data Cleaning

df.isnull().sum()
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.head()
```

```python
# step5 : here, we will understand the target variables 
# G1 - First period grade
# G2 - Second period grade
# G3 - Final grade(Main Target)

df[['G1','G2','G3']].head()
```

```python
# step6 : create total_score and average_score 

df['total_score'] = df['G1'] + df['G2'] + df['G3']
df['average_score'] = df['total_score'] / 3
```

```python
# step7 : categorical feature distribution (we will check parent's education)
```

```python
sns.countplot(x='Medu', data=df)
plt.title("Mother's Education Level Distribution")
plt.show()
```

```python
sns.countplot(x='Fedu', data=df)
plt.title("Father's Education Level Distribution")
plt.show()
```

```python
# step8 : Parent's Education vs Performance
```

```python
sns.boxplot(x='Medu', y='average_score', data=df)
plt.title("Mother's Education vs Student Performance")
plt.show()
```

```python
sns.boxplot(x='Fedu', y='average_score', data=df)
plt.title("Father's Education vs Student Performance")
plt.show()
```

```python
# step9 : we will check Family size impact

sns.boxplot(x='famsize', y='average_score', data=df)
plt.title("Family Size vs Student Performance")
plt.show()
```

```python
# step10 : Parent's Marital Status

sns.boxplot(x='Pstatus', y='average_score', data=df)
plt.title("Parents' Marital Status vs Performance")
plt.show()
```

```python
# step11 : Family Relationship Quality

sns.boxplot(x='famrel', y='average_score', data=df)
plt.title("Family Relationship Quality vs Performance")
plt.show()
```

```python
# step12: Alcohol Consumption Analysis
```

```python
# here we will see Weekday Alcohol Consumption

sns.boxplot(x='Dalc', y='average_score', data=df)
plt.title("Weekday Alcohol Consumption vs Performance")
plt.show()
```

```python
# here, Weekend Alcohol Consumption

sns.boxplot(x='Walc', y='average_score', data=df)
plt.title("Weekend Alcohol Consumption vs Performance")
plt.show()
```

```python
# step13 : Study Time vs Performance

sns.boxplot(x='studytime', y='average_score', data=df)
plt.title("Study Time vs Student Performance")
plt.show()
```

```python
# step14 : Gender vs Performance

sns.boxplot(x='sex', y='average_score', data=df)
plt.title("Gender vs Academic Performance")
plt.show()
```

```python
# step15 : here we'll do the Correlation Analysis

corr = df[['G1','G2','G3','studytime','failures','Dalc','Walc','average_score']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
